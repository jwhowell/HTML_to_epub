```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://wesmckinney.com/book/"
#url = "https://wesmckinney.com/book/advanced-numpyipython.html"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html")
soup
```

```python
side_bar = soup.find_all(class_ = "sidebar-menu-container")
side_bar
```

```python
link = soup.find("a", href=True)
link["href"]
```

```python
BASE = "https://www.typescriptlang.org"
with open("side_bar.xml", "r") as file:
    text = file.read()
soup = BeautifulSoup(text)
output = soup.find_all("a", href=True)
links = []
from urllib.parse import urljoin
for o in output:
    href = o.get("href")
    if not href or href.startswith(("#", "javascript:")):
        continue
    title = o.get_text(strip=True)
    links.append((title, urljoin(BASE, href)))
print(links)

```

```python
from bs4 import BeautifulSoup
import re
BASE = "https://natureofcode.com"
with open("nature_of_code_sidebar.html", "r") as file:
    text = file.read()
soup = BeautifulSoup(text)
output = soup.find_all("a", href=True)
links = []
from urllib.parse import urljoin
for o in output:
    href = o.get("href")
    if not href or href.startswith(("#", "javascript:")):
        continue
    title = o.get_text(strip=True)
    title = re.sub(r'^\d+', '', title)
    links.append((title, urljoin(BASE, href)))
print(links)

```
